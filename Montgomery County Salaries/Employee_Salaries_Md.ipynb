```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Read Excel File

```python
df = pd.read_csv("Employee_Salaries_-_2023.csv")
print(df.head(5))
```

```python
mod_df = df.loc[:, ["Department", "Department_Name", "Gender", "Base_Salary", "Overtime_Pay", "Longevity_Pay"]]
mod_df
```

Male Salaries (Average)

```python
male_sal = mod_df[mod_df["Gender"] == "M"]
male_sal
```

```python
male_sal_avg = male_sal.groupby(["Department"]).mean(numeric_only=True)
male_sal_avg
```

```python
male_x = [dep for dep, df in male_sal_avg.groupby("Department")]
print(male_x)
```

```python
male_base = male_sal_avg.loc[:, "Base_Salary"]
male_over = male_sal_avg.loc[:, "Overtime_Pay"]
male_long = male_sal_avg.loc[:, "Longevity_Pay"]
```

```python
plt.plot(male_x, male_base, label =  "Base Salary", marker = "o")
plt.plot(male_x, male_over, label =  "Overtime Pay", marker = "o")
plt.plot(male_x, male_long, label =  "Longevity Pay", marker = "o")

plt.title("Avg. Pay for Male Gov. Employees Montgomery County, MD 2023")
plt.xlabel("Departments")
plt.ylabel("Average Pay (US Dollars)")

plt.xticks(rotation = 90, ha = "center")
plt.grid(True, which ="both", color= "gray")

plt.legend(title="Salary Information", loc = "upper right")
plt.show()
```

Female Salaries (Average)

```python
female_sal = mod_df[mod_df["Gender"] == "F"]
female_sal
```

```python
female_sal_avg = female_sal.groupby(["Department"]).mean(numeric_only=True)
female_sal_avg
```

```python
female_x = [dep for dep, df in female_sal_avg.groupby("Department")]

female_base = female_sal_avg.loc[:, "Base_Salary"]
female_over = female_sal_avg.loc[:, "Overtime_Pay"]
female_long = female_sal_avg.loc[:, "Longevity_Pay"]

plt.plot(female_x, female_base, label =  "Base Salary", marker = "o")
plt.plot(female_x, female_over, label =  "Overtime Pay", marker = "o")
plt.plot(female_x, female_long, label =  "Longevity Pay", marker = "o")

plt.title("Avg. Pay for Female Gov. Employees Montgomery County, MD 2023")
plt.xlabel("Departments")
plt.ylabel("Average Pay (US Dollars)")

plt.xticks(rotation = 90, ha = "center")
plt.grid(True, which ="both", color= "gray")

plt.legend(title="Salary Information", loc = "upper right")
plt.show()
```
